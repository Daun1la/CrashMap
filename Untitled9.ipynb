```python
from google.colab import files
import numpy as np
import matplotlib.pyplot as plt
import random
from collections import Counter
from scipy.stats import uniform
from scipy.stats import norm
from scipy.stats import chisquare
import math
from statistics import mean
from statistics import median

uploaded = files.upload()

with open('Waveform1.txt', 'r') as f:
    content = f.read()

#перевод исходного списка в float список с удалением пустого элемента, и заменой первого на ноль
str_arr = content.split('\n')
float_list = []
for item in str_arr[1:]:
    if item != '':
        float_list.append(float(item))
float_list.insert(0, 0)

impulse = np.array(float_list)
shag = 32767
```

```python
# Равномерное распределение без наложения и с амплитудным распределением


# Задание диапазона амплитуд
min_amplitude = 0.2
max_amplitude = 0.7

# Количество подобных импульсов для генерации
num_impulses = 10000

# Создание массива амплитуд с равномерным распределением
amplitudes = uniform.rvs(loc=min_amplitude, scale=max_amplitude - min_amplitude, size=num_impulses)

# Генерация подобных импульсов с заданными амплитудами
similar_impulses = amplitudes[:, np.newaxis] * impulse / max(float_list) #здесь делим на максимальное значение, чтобы привести амплитуду импульса к единице и в дальнейшем эту единицу умножать на сгенерирированный массив

# Получение количества импульсов и отсчетов
num_impulses, num_samples = similar_impulses.shape

# Объединение импульсов в один массив
# тут пока не понял как последовательно выводить импульсы когда они в виде ndaray, поэтому перевожу заново в обычный список
combined_data = np.hstack(similar_impulses)

# Создание оси времени
time_axis = np.arange(num_samples * num_impulses) * 4  # Предполагая, что каждый отсчет равен 4 временным единицам

# Визуализация объединенного массива
fig = plt.figure(figsize=(20, 6))
plt.plot(time_axis, combined_data)
plt.title('Импульсы')
plt.xlabel('Время, нс')
plt.ylabel('Амплитуда,В')
plt.show()

#гисторграмма для равномерного распределения
max_arr = []

for y in similar_impulses:
  max_arr.append(y.max())

# Создание гистограммы
bin = 0.05
bins = np.arange(min_amplitude, max_amplitude, bin) # в этой строке задается диапозон амплитуд, показанный на графике и ширина бина, в котором будут считаться значения
hist, edges = np.histogram(max_arr, bins=bins)

# Настройка графика
fig = plt.figure(figsize=(10,5))
plt.bar(bins[:-1], hist, width=bin*0.9, align='edge', edgecolor='black') # здесь параметр width меняет ширину бина на графике
plt.title('Распределение амплитуд')
plt.xlabel('Значение, В')
plt.ylabel('Частота')

# вывод количества насчитанных амплитуд
for i in range(len(hist)):
    plt.text(edges[i], hist[i], str(hist[i]), ha='left', va='bottom')

plt.grid(True)

# проверка на критерий Пирсона (не особо понимаю как работает)
# Расчет ожидаемого количества значений в каждом бине для нормального распределения
expected_counts = len(max_arr) * np.diff(bins) * norm.pdf(bins[:-1], loc=min_amplitude, scale=max_amplitude - min_amplitude)

# Нормализация наблюдаемых и ожидаемых частот
observed_freq = hist / np.sum(hist)
expected_freq = expected_counts / np.sum(expected_counts)

# Расчет статистики критерия Пирсона и p-значений
chi2, p_value = chisquare(observed_freq, expected_freq)

print("Статистика критерия Пирсона:", chi2)
print("p-значение:", p_value)
```

```python
# Генерация временного распределения по формуле
# t = -(count_rate)^(-1)*log10(r), где r - случайное число, r(0;1]
times = []
N_times = 10000
count_rate = 1e+07
sample_value = 4
# создаем массив из N элементов, по формуле выше
for t in range(N_times):
    times.append(-math.log10(random.uniform(1e-20, 1))/(count_rate*sample_value)*1e+06*4) #мкс

# Время, генерируемое по этой формуле физическое - 1 нс эквивалентна 1 нс
# Но время, в котором описывается импульс имеет цену деления в 4нс
# Получается, что 4нс в массиве times должны быть переведены в 1 отсчет
# И уже в таком виде учитываться в сложении импульсов
# диапазон для подсчета
bin_range = (0, 40)
bins = 20

# Подсчет количества значений в диапазоне
hist, bin_edges = np.histogram(times, bins, range=bin_range)

fig = plt.figure(figsize=(10, 5))
plt.hist(times, bins, range=bin_range)
plt.grid(True)
# Вывод значений количества временных интервалов над каждым бином
bin_start = bin_edges[0]
bin_end = bin_edges[1]
for i in range(len(hist)):
    bin_center = (bin_edges[i] + bin_edges[i+1]) / 2
    count = hist[i]
    plt.text(bin_center, count, str(count), ha='left', va='baseline')
print("Ширина бина: " + str((bin_end-bin_start))+ ' мкс')
print(mean(times))
plt.title('Временное распределение')
plt.ylabel('Количество')
plt.xlabel('Время,мкс')
plt.show()
```

```python
def lil_imp():
  new_impulse = float_list.copy()
  const = random.uniform(5450/shag, (5450+100)/shag)
  new_impulse = np.array(new_impulse)
  return new_impulse*const/max(float_list)

def avl_imp():
  new_impulse = float_list.copy()
  const = random.uniform(6450/shag, (6450+100)/shag)
  new_impulse = np.array(new_impulse)
  return new_impulse*const/max(float_list)

def avr_imp():
  new_impulse = float_list.copy()
  const = random.uniform(7950/shag, (7950+100)/shag)
  new_impulse = np.array(new_impulse)
  return new_impulse*const/max(float_list)

def big_imp():
  new_impulse = float_list.copy()
  const = random.uniform(10950/shag, (10950+100)/shag)
  new_impulse = np.array(new_impulse)
  return new_impulse*const/max(float_list)

def big1_imp():
  new_impulse = float_list.copy()
  const = random.uniform(14950/shag, (14950+100)/shag)
  new_impulse = np.array(new_impulse)
  return new_impulse*const/max(float_list)

def big2_imp():
  new_impulse = float_list.copy()
  const = random.uniform(24950/shag, (24950+100)/shag)
  new_impulse = np.array(new_impulse)
  return new_impulse*const/max(float_list)

def big3_imp():
  new_impulse = float_list.copy()
  const = random.uniform(29950/shag, (29950+100)/shag)
  new_impulse = np.array(new_impulse)
  return new_impulse*const/max(float_list)
```
